```python
import socket
import keyboard

HOST = '0.0.0.0'
PORT = 5000

server = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
server.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)  # allows restart without waiting
server.bind((HOST, PORT))
server.listen(1)

print(f"Waiting for connection on port {PORT}...")
client, addr = server.accept()
print(f"Connected: {addr}")

while True:
    try:
        data = client.recv(1024).decode().strip()  # ← .strip() fixes the \n issue
        if not data:
            break
        print(f"Received: '{data}'")
        keyboard.press_and_release(data)
    except Exception as e:
        print(f"Error: {e}")
        break

client.close()
server.close()
print("Connection closed.")
```
